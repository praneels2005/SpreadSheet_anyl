```python
import pandas as pd
import numpy as np
import locale
import ast
import datetime
from openpyxl import Workbook,load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import get_column_letter
import numpy as np
from openpyxl.styles import PatternFill

locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
red_fill = PatternFill(start_color='FFFF0000', end_color='FFFF0000', fill_type='solid')
green_fill = PatternFill(start_color='FF00FF00', end_color='FF00FF00', fill_type='solid')
yellow_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
```

```python
Months = {1:"Jan",
         2:"Feb",
         3:"Mar",
         4:"Apr",
         5:"May",
         6:"Jun",
         7:"Jul",
         8:"Aug",
         9: "Sep",
         10:"Oct",
         11:"Nov",
         12:"Dec"}
```

```python
data = pd.read_csv('Mahe-Challenge-Copy - ChallengeCopy.csv')
data
```

```python
data['Trade Enter'] = [f'{Months.get(int(y[1]))} {y[2]}, {y[0]}' for date in data['Trade Enter'] if (y:=date.split('-'))]
data['Trade Exit'] = [f'{Months.get(int(y[1]))} {y[2]}, {y[0]}' for date in data['Trade Exit'] if (y:=date.split('-'))]
data['Expiry Date'] = [f'{Months.get(int(y[1]))} {y[2]}, {y[0]}' for date in data['Expiry Date'] if (y:=date.split('-'))]
data['Strike'] = [locale.currency(float(price)) for price in data['Strike']]
data
```

```python
data['Realized Gain $'] = [locale.currency(float(price)) if price>0 else f'({locale.currency(float(price)*-1)})'for price in data['Realized Gain $']]
```

```python
data = data.rename(columns={'Trade #': 'Index', 'Expiry Date':'Exp Date', 'Realized Gain $': '$ Gain'})
data = data.loc[:,['Index', 'Ticker', 'Trade Enter', 'Strike', 'C/P','Exp Date', '# of Contracts', 'Trade Exit', '$ Gain']]
data
```

```python
data.to_csv('Mahe-Challenge.csv',index=False)
```
